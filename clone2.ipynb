```python
# Clone and Explore Repository
This notebook will clone the GitHub repository `https://github.com/viberverse/viberverse.github.io.git` into the workspace root and allow us to interact with its files.
</VSCode.Cell>
<VSCode.Cell language="python">
# Section 1: Clone the GitHub Repository
import os
import subprocess

repo_url = "https://github.com/viberverse/viberverse.github.io.git"
clone_dir = os.getcwd()

# attempt to clone with git via subprocess
try:
    subprocess.run(["git", "clone", repo_url, clone_dir], check=True)
    print("Repository cloned successfully.")
except subprocess.CalledProcessError as e:
    print("Git clone returned an error (likely because target already exists):", e)

# Alternatively, could use GitPython if available
</VSCode.Cell>
<VSCode.Cell language="markdown">
## Section 2: List Cloned Files in the Root Directory
Here we'll list the files and folders now present in the root directory after cloning.
</VSCode.Cell>
<VSCode.Cell language="python">
import pathlib

root_path = pathlib.Path(clone_dir)
files = list(root_path.iterdir())
print(f"Total entries in root: {len(files)}")
for f in files:
    print(f.name)
</VSCode.Cell>
<VSCode.Cell language="markdown">
## Section 3: Read and Display a File from the Repository
We'll open a file (e.g., `index.html` if it exists) and display its contents. Replace the filename if needed.
</VSCode.Cell>
<VSCode.Cell language="python">
file_to_read = root_path / "index.html"

if file_to_read.exists():
    with file_to_read.open() as fh:
        content = fh.read()
    print(content[:1000])  # print first 1000 characters
else:
    print(f"{file_to_read} does not exist in repository.")
```
